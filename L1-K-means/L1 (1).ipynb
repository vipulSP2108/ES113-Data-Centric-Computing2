# Que 1. K-Means
Use sklearn.load_digits() to load the digits datasetImplement the k-means algorithm (from scratch) on this dataset.

1. Use k = 10

2. Pick the initial centers to be uniform at random from all the points

3. Run for 100 iterations Plot how the objective value decreases with number of iterations. Use imshow() to show the images corresponding to the centers of the clusters

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()  # for plot styling
import numpy as np
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix

digits=load_digits()
k=10
iterations=100
```

```python
print('fig 1')
def k_mean(digits,k,iterations):
  np.random.seed(100)
  centers=digits.data[np.random.choice(digits.data.shape[0],k,replace=False)]
  mat=np.zeros(iterations)
  for i in range(0,iterations):
    dist=np.linalg.norm(digits.data[:,np.newaxis,:]-centers,axis=2)
    labels=np.argmin(dist, axis=1)
    for j in range(k):
      centers[j]=np.mean(digits.data[labels==j],axis=0)
    mat[i]=np.sum(np.min(dist,axis=1))
    i=i+1
  return mat,centers

val,centers=k_mean(digits,k,iterations)

plt.figure(figsize=(5,5))
plt.plot(val)
plt.xlabel('Iterations')
plt.ylabel('Objective Value')
plt.title('K-Means (Digits Dataset)')
plt.show()

print('fig 2')
fig,ax=plt.subplots(1,10,figsize=(12,4))
for i,axi in enumerate(ax.flat):
  axi.set(xticks=[],yticks=[])
  axi.imshow(centers[i].reshape(8,8),cmap='binary')
  axi.set_xlabel(f'Cluster {i}')
plt.show()
```
